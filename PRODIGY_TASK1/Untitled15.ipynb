```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    r"C:\Prodigy\API_SP.POP.TOTL_DS2_en_csv_v2_174326\API_SP.POP.TOTL_DS2_en_csv_v2_174326.csv",
    skiprows=4
)
```

```python
data_2022 = df[['Country Name', '2022']].dropna()
top10 = data_2022.sort_values(by='2022', ascending=False).head(10)
population_all = df['2022'].dropna()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18,6))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18,6))

# Bar chart
sns.barplot(x='Country Name', y='2022', data=top10, palette="viridis", ax=axes[0])
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right')
axes[0].set_xlabel("Country")
axes[0].set_ylabel("Population")
axes[0].set_title("Top 10 Countries by Population (2022)")

# Histogram
population_all = df['2022'].dropna()
sns.histplot(population_all, bins=20, kde=True, color='skyblue', ax=axes[1])
axes[1].set_xlabel("Population")
axes[1].set_ylabel("Number of Countries")
axes[1].set_title("Population Distribution Across Countries (2022)")

plt.tight_layout()
plt.show()
```
